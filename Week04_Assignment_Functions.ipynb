# Q 1. What is the difference between a function and a method in Python?

# Function
  - Function is a block of reusable code that can perform specific tasks.
  - it is defined independently
  - called directly using name.
  - does not reuire self
  

```python
# eg
def sum(a, b):
  return a + b
sum(5, 56)
```

# Method
   - method is a function associated with object and operates on object.
   - it is defined insided a class and belongs to an object.
   - called using object of a class.
   - require self as first parameter.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def greet(self):
    return f"hello i'm {self.name} and i'm {self.age}."

person1 = Person("asad", 23)
print(person1.greet())
```

# 4. Function with Variable-Length Arguments (*args and **kwargs)
  - *args: Allows multiple positional arguments.
  - **kwargs: Allows multiple keyword arguments.

```python
def sum_all(*numbers):  # *args collects multiple values
    return sum(numbers)

def info(**details):  # **kwargs collects multiple key-value pairs
    for key, value in details.items():
        print(f"{key}: {value}")
```

# Q 6. Explain the concept of generators in Python and how they are defined.

- A generator is a special type of itrator that yield value lazyly insted of returning all at a time it gives only one and pause the execution. This makes it memory efficient for handling the data.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield one value at a time
        count += 1

# Using the generator
gen = count_up_to(5)

print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # output : 3
```

- **Support infinite sequence** - can generate values infinitely without memory usage.

- **Improves performance** - useful for large dataaset

# Q 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even(number):
  total = 0
  for i in number:
    if (i % 2 == 0):
      total = total + i
  return total
number_list = [5,6,8,7,9,10,62,35]
print(sum_even(number_list))
```

```python
def Prime(number):
  if (number < 2 ):
    return False
  for i in range(2, number):
    if number % i == 0:
      return False
  return True

for num in range(1, 201):
  if Prime(num):
    print(num, end = " ")

```

# Q 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fibbo(number):
  a, b = 0 , 1
  for _ in range(number):
    print(a, end = " ")
    a, b = b, a + b
fibbo(5)
```

# Q 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(exponent):
  for i in range(exponent + 1):
    yield 2 ** i

for power in power_of_two(5):
  print(power, end = " ")
```

# Q 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
with open("/content/sample.txt", "w") as file:
    file.write("Hello, this is a test file!\nSecond line here.")
```

```python
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()  # Yield each line without trailing newline

# Example usage
for line in read_file_line_by_line("/content/sample.txt"):
    print(line)  # Process each line
```
