# Q14. How does the “in” keyword work differently for lists and dictionaries?

- list - When the in keyword is used with a list, it checks if the specified element exists within the list. It searches through the list and returns True if the element is found, and False if it is not.
- dictionary - When the in keyword is used with a dictionary, it checks if the specified key exists in the dictionary, not the value. By default, it looks for the key in the dictionary's keys.


```python
# list
name = ["ajay", "rohan", "sunil", "vijay"]
"ajay" in name
```

```python
# dictionary
my_dict = {"a": 1, "b": 2, "c": 3}
print("a" in my_dict)
print("d" in my_dict)
```

# Q17. Describe the time complexity of accessing elements in a dictionary?

- Average time complexity:
- The time complexity of accessing elements in a dictionary in Python is O(1) on average. This means that, in most cases, retrieving the value associated with a specific key is done in constant time, regardless of the size of the dictionary.
- Worst-Case Time Complexity:
- In the worst case, dictionary access could be O(n) if there are hash collisions that are handled poorly. A hash collision occurs when two keys have the same hash value, and the dictionary needs to handle this collision by chaining or probing. However, modern Python implementations minimize this risk by using techniques like open addressing and resizeable hash tables, making hash collisions rare.

- Thus, in practice, O(1) is the expected time complexity for most dictionary operations, including accessing elements, inserting new items, and deleting items.

# Q20. Explain the difference between a list and a dictionary in terms of data retrieval.


- List
    - Ordered collection of elements indexed by integers.
    - Access by index (e.g., my_list[0]).
    - O(1) for accessing by index.
    - Preserves the order of insertion.
    - Data is accessed by position (index).
    - Best for ordered collections.
- Dictionary
    - Unordered collection of key-value pairs.
    - Access by key (e.g., my_dict['key']).
    - O(1) average for retrieving by key.
    - Insertion order is preserved (Python 3.7+), but not guaranteed in earlier versions.
    - Data is accessed by a unique key.
    - Best for fast lookups and key-value associations.
